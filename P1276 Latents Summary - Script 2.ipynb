```python
import ipywidgets as widgets
from matplotlib import pyplot as plt
import matplotlib
import numpy
import math
import datetime
import imageio
import glob
from PIL import Image
import pandas as pd
from pathlib import Path
from time import time
import sys
import scipy.ndimage
from scipy import stats 
import shutil
import os
from re import search
import copy
from sklearn.decomposition import NMF
from numpy.random import RandomState
import calendar
from os.path import exists
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
from PIL import Image, ImageTk
import time
import tkinter as tk
from scipy.stats import rankdata
import matplotlib.cm as cm
import matplotlib.pylab as pyl
#import PyUber, PyUber.client
import threading

##############################################
# User Defined Variables
##############################################
ActiveDate = '20230505' # Format: YYYYMMDD
InputFile = 'C:/Users/DSMIADAK/OneDrive - Intel Corporation/76 Fleet/20230505/X Edge/output_file2_20230505.csv'
##############################################

def convert(seconds):
    seconds = seconds % (24 * 3600)
    hour = seconds // 3600
    seconds %= 3600
    minutes = seconds // 60
    seconds %= 60
     
    return "%d:%02d:%02d" % (hour, minutes, seconds)

startTime = time.time()
print("Program Started...Please wait")

OutputFile = os.path.dirname(InputFile) + '/output_file3_' + ActiveDate + '.csv' 

# ------------- Extract LOT data as done previously ----------------- #
df_csv1 = None
df_csv1 = pd.read_csv(InputFile, low_memory=False)
df = pd.DataFrame()
df = pd.DataFrame(df_csv1)

## Get Access to all images across all workweeks
df2 = None
df2 = df
df2 = df2.drop_duplicates()
df2 = df2.dropna()

df_tool = df2['Process Equipment'].drop_duplicates()
# print(len(df_tool))
df_Ins = df2['Inspection Date'].drop_duplicates()
df_Start = df2['Link Start Time'].drop_duplicates()
df_Lot = df2['Lot'].drop_duplicates()
df_Slot = df2['Slot'].drop_duplicates()
df_Location = df2['Location'].drop_duplicates()
df_Zone = df2['Zone'].drop_duplicates()

Loc_Zone = [('Center Count', 'Center Count'),
            ('Edge Count', 'Edge Count - Non-Film Pulling Region'),
            ('Edge Count', 'Edge Count - Film Pulling Region'),
            ('Extreme Edge Count', 'Extreme Edge Count - Non-Film Pulling Region'),
            ('Extreme Edge Count', 'Extreme Edge Count - Film Pulling Region')]
df_Loc_Zone = pd.DataFrame(Loc_Zone, columns =['Location', 'Zone'])
C_count = [('0')]
df_C_count = pd.DataFrame(C_count, columns =['Count'])
line_count=0
i = 1

for tool in df_tool:
    print('Now processing data for Entity: ' + str(tool) + ' (Tool ' + str(i) + ' of ' + str(len(df_tool)) + ')')
    for Ins in df_Ins:
        print('*** Processing: ' + str(Ins))
        df_1=pd.DataFrame()
        df_1=copy.copy(df2.loc[((df2['Process Equipment']==tool) & (df2['Inspection Date']==Ins))])
        if df_1.empty ==  False:
            for Start in df_Start:
                df_2=pd.DataFrame()
                df_2=copy.copy(df_1.loc[(df_1['Link Start Time']==Start)])
                if df_2.empty ==  False:
                    for Lot in df_Lot:
                        df_3=pd.DataFrame()
                        df_3=copy.copy(df_2.loc[(df_2['Lot']==Lot)])
                        for Slot in df_Slot:
                            df_4=pd.DataFrame()
                            df_4=copy.copy(df_3.loc[(df_3['Slot']==Slot)])
                            if df_4.empty==False:
                                ir=None
                                Bulk_DF=pd.DataFrame()
                                for ir in range(0,numpy.shape(df_Loc_Zone)[0]):
                                    df_5=pd.DataFrame()
                                    df_5=copy.copy(df_4.loc[((df_4['Location']==df_Loc_Zone.loc[ir]['Location']) & 
                                                             (df_4['Zone']==df_Loc_Zone.loc[ir]['Zone']))])
                                    if df_5.empty:
                                        #Append the relevant datasets
                                        df_5_add=pd.DataFrame()
                                        df_5_add=pd.concat([pd.DataFrame([tool]),pd.DataFrame([Ins]),
                                                            pd.DataFrame([Start]),pd.DataFrame([Lot]),
                                                            pd.DataFrame([Slot]),
                                                            pd.DataFrame([df_Loc_Zone.loc[ir]['Location']]),
                                                            pd.DataFrame([df_Loc_Zone.loc[ir]['Zone']]),
                                                            pd.DataFrame(['0'])],axis=1)
                                        df_5_add.columns = df.columns
                                    elif df_5.empty ==  False:
                                        df_5_add=pd.DataFrame()
                                        df_5_add=copy.copy(df_5)
                                    
                                    Bulk_DF=pd.concat([Bulk_DF,df_5_add])
                                    Bulk_DF.columns = df.columns
                                    
                                if line_count != 0:
                                    Bulk_DF.to_csv(OutputFile,header=False,index=False,mode='a')
                                else:
                                    Bulk_DF.to_csv(OutputFile,header=True,index=False,mode='a')
                                    
                                line_count=line_count+1
    i = i + 1
executionTime = (time.time() - startTime)
print('Execution time in seconds: ' + str(convert(executionTime)))                           
```
